```python
import pandas as pd

# Step 1: Assign correct column names
cols = ['age', 'workclass', 'fnlwgt', 'education', 'education-num',
        'marital-status', 'occupation', 'relationship', 'race', 'sex',
        'capital-gain', 'capital-loss', 'hours-per-week',
        'native-country', 'income']

# Step 2: Load dataset (IMPORTANT FIX)
df = pd.read_csv("adult.csv", names=cols, skipinitialspace=True)

# Optional check
print(df.head())
print(df.columns)
```

```python
# s. Create subsets (Country, Sex, Race)
# ----------------------------------------------------
india_data = df[df['native-country'] == 'India']
male_data = df[df['sex'] == 'Male']
white_data = df[df['race'] == 'White']

print("\nIndia Data:")
print(india_data.head())

print("\nMale Data:")
print(male_data.head())

print("\nWhite Data:")
print(white_data.head())
```

```python
# t. Merge two subsets
# ----------------------------------------------------
merged_data = pd.concat([india_data, male_data])

print("\nMerged Data:")
print(merged_data.head())
```

```python
# u. Sort data (using age)
# ----------------------------------------------------
sorted_data = df.sort_values(by='age', ascending=False)

print("\nSorted Data:")
print(sorted_data[['age', 'sex', 'native-country']].head())
```

```python
# v. Transpose data
# ----------------------------------------------------
transposed_data = df.head().T

print("\nTransposed Data:")
print(transposed_data)
```

```python
# w. Melt data (Wide → Long)
# ----------------------------------------------------
melted_data = pd.melt(df,
                     id_vars=['age', 'sex'],
                     value_vars=['race', 'native-country'],
                     var_name='Category',
                     value_name='Value')

print("\nMelted Data:")
print(melted_data.head())
```

```python
# x. Cast data (Long → Wide)
# ----------------------------------------------------
wide_data = melted_data.pivot_table(index=['age', 'sex'],
                                    columns='Category',
                                    values='Value',
                                    aggfunc='first')

print("\nWide Data:")
print(wide_data.head())
```
