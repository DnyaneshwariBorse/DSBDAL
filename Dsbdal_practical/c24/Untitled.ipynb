```python
# Step 1 : Import Libraries
import pandas as pd
```

```python
# Step 2 : Load Dataset
# Make sure forestfires.csv is in same folder
df = pd.read_csv("forestfires.csv")
```

```python
# Display first 5 rows
print("Original Dataset:\n")
print(df.head())
```

```python
# =========================================================
# a. Create data subsets by making classes for area affected
# =========================================================

# We will create 3 classes using 'area' column

# NotAffected       -> area = 0
# PartiallyAffected -> area > 0 and area <= 50
# MostlyAffected    -> area > 50

not_affected = df[df['area'] == 0]

partially_affected = df[(df['area'] > 0) & (df['area'] <= 50)]

mostly_affected = df[df['area'] > 50]

print("\n\nNot Affected Data:\n")
print(not_affected.head())

print("\n\nPartially Affected Data:\n")
print(partially_affected.head())

print("\n\nMostly Affected Data:\n")
print(mostly_affected.head())
```

```python

# ==================================
# b. Merge two subsets
# ==================================

# Merging NotAffected and PartiallyAffected

merged_data = pd.concat([not_affected, partially_affected])

print("\n\nMerged Dataset:\n")
print(merged_data.head())
```

```python
# ==================================
# c. Sort Data
# ==================================

# Sorting using temperature, wind and area

sorted_data = df.sort_values(
    by=['temp', 'wind', 'area'],
    ascending=[True, True, True]
)

print("\n\nSorted Dataset:\n")
print(sorted_data.head())
```

```python
# ==================================
# d. Transposing Data
# ==================================

transpose_data = df.transpose()

print("\n\nTransposed Dataset:\n")
print(transpose_data.head())
```

```python
# ==================================
# e. Melting Data to Long Format
# ==================================

# Converting temp, RH, wind columns to long format

melted_data = pd.melt(
    df,
    id_vars=['month'],
    value_vars=['temp', 'RH', 'wind'],
    var_name='Weather_Type',
    value_name='Value'
)

print("\n\nMelted Data:\n")
print(melted_data.head())
```

```python
# ==================================
# f. Casting Data to Wide Format
# ==================================

# Convert melted data back to wide format

wide_data = melted_data.pivot_table(
    index='month',
    columns='Weather_Type',
    values='Value',
    aggfunc='mean'
)

print("\n\nWide Format Data:\n")
print(wide_data.head())
```
