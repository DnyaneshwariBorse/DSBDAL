```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ==========================================
# COLUMN NAMES
# ==========================================

columns = [
    'age',
    'workclass',
    'fnlwgt',
    'education',
    'education-num',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'capital-gain',
    'capital-loss',
    'hours-per-week',
    'native-country',
    'income'
]

# ==========================================
# LOAD DATASET
# ==========================================

df = pd.read_csv(
    "adult.csv",
    header=None,
    names=columns
)

# ==========================================
# DISPLAY FIRST 5 ROWS
# ==========================================

print(df.head())
```

```python
# ==========================================
# DATA CLEANING
# ==========================================

# Replace missing values represented by '?'
df.replace('?', pd.NA, inplace=True)

# Remove missing values
df.dropna(inplace=True)
```

```python
# ==========================================
# 1. HISTOGRAM
# ==========================================
# Objective:
# To understand distribution of age

plt.figure(figsize=(8,5))

plt.hist(df['age'], bins=10)

plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
# ==========================================
# 2. DOT PLOT
# ==========================================
# Objective:
# To visualize spread of hours-per-week values

plt.figure(figsize=(10,4))

sns.stripplot(x=df['hours-per-week'])

plt.title("Dot Plot of Hours Per Week")

plt.show()
```

```python
# ==========================================
# 3. BAR PLOT
# ==========================================
# Objective:
# To compare workclass frequencies

plt.figure(figsize=(10,5))

df['workclass'].value_counts().plot(kind='bar')

plt.title("Bar Plot of Workclass")
plt.xlabel("Workclass")
plt.ylabel("Count")

plt.show()
```

```python
# ==========================================
# 4. LINE CHART
# ==========================================
# Objective:
# To study trend between age and hours-per-week

df_sorted = df.sort_values(by='age')

plt.figure(figsize=(10,5))

plt.plot(
    df_sorted['age'],
    df_sorted['hours-per-week']
)

plt.title("Line Chart: Age vs Hours Per Week")
plt.xlabel("Age")
plt.ylabel("Hours Per Week")

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
# Left graph: Histogram
sns.histplot(df['age'], color='lightblue', ax=ax1)
ax1.set_title("Histogram")
# Right graph: Box Plot with Scatter Plot added on top
sns.boxplot(x=df['age'], color='lightgreen', ax=ax2)
sns.stripplot(x=df.sample(500)['age'], color='red', alpha=0.5, ax=ax2) 
ax2.set_title("Box Plot + Scatter Plot")
plt.show()
```
