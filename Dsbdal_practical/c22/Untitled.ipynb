```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================
# COLUMN NAMES
# ==========================================

columns = [
    'age',
    'workclass',
    'fnlwgt',
    'education',
    'education-num',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'capital-gain',
    'capital-loss',
    'hours-per-week',
    'native-country',
    'income'
]

# ==========================================
# LOAD DATASET
# ==========================================

df = pd.read_csv(
    "adult.csv",
    header=None,
    names=columns
)

# ==========================================
# DISPLAY FIRST 5 ROWS
# ==========================================

print(df.head())
```

```python
# ==========================================
# DATA CLEANING
# ==========================================

# Replace missing values represented by '?'
df.replace('?', pd.NA, inplace=True)

# Remove missing values
df.dropna(inplace=True)
```

```python
# ==========================================
# 1. HISTOGRAM
# ==========================================
# Objective:
# To understand distribution of age

plt.figure(figsize=(8,5))

plt.hist(df['age'], bins=10)

plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
# ==========================================
# 2. PIE CHART
# ==========================================
# Objective:
# To show proportion of income categories

income_counts = df['income'].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    income_counts,
    labels=income_counts.index,
    autopct='%1.1f%%'
)

plt.title("Income Distribution")

plt.show()
```

```python
# ==========================================
# 3. BOX PLOT
# ==========================================
# Objective:
# To detect outliers in hours-per-week

plt.figure(figsize=(8,5))

sns.boxplot(x=df['hours-per-week'])

plt.title("Box Plot of Hours Per Week")

plt.show()
```

```python
# ==========================================
# 4. SCATTER PLOT
# ==========================================
# Objective:
# To study relationship between age and
# hours-per-week

plt.figure(figsize=(8,5))

plt.scatter(
    df['age'],
    df['hours-per-week']
)

plt.title("Scatter Plot: Age vs Hours Per Week")
plt.xlabel("Age")
plt.ylabel("Hours Per Week")

plt.show()
```

```python
# ==========================================
# 5. ADD BOXPLOTS TO SCATTERPLOT
# ==========================================
# Objective:
# To visualize relationship between age and
# hours-per-week along with distributions
# and outliers

sns.jointplot(
    x='age',
    y='hours-per-week',
    data=df,
    kind='scatter'
)

plt.show()

# ==========================================
# END
# ==========================================
```
