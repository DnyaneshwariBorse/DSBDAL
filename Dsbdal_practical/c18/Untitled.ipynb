```python
# ==========================================
# HEART DISEASE DATASET VISUALIZATION
# ==========================================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ==========================================
# COLUMN NAMES
# ==========================================

columns = [
    'age', 'sex', 'cp', 'trestbps', 'chol',
    'fbs', 'restecg', 'thalach', 'exang',
    'oldpeak', 'slope', 'ca', 'thal', 'target'
]

# ==========================================
# LOAD DATASET
# ==========================================

df = pd.read_csv("Cleavland.csv", header=None, names=columns)
```

```python
# ==========================================
# DISPLAY FIRST 5 ROWS
# ==========================================

print(df.head())
```

```python
# ==========================================
# DATA CLEANING
# ==========================================

# Replace missing values (?) with NaN
df.replace('?', pd.NA, inplace=True)

# Remove missing values
df.dropna(inplace=True)

# Convert all columns to numeric
for col in df.columns:
    df[col] = pd.to_numeric(df[col])
```

```python
# ==========================================
# 1. HISTOGRAM
# ==========================================
# Objective:
# To understand distribution of age

plt.figure(figsize=(8,5))

plt.hist(df['age'], bins=10)

plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
# ==========================================
# 2. PIE CHART
# ==========================================
# Objective:
# To show proportion of heart disease cases

target_counts = df['target'].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    target_counts,
    labels=target_counts.index,
    autopct='%1.1f%%'
)
# in pie target_counts.index is  used to giving indexing to each pie
#autopct is auto percent 

plt.title("Heart Disease Distribution")

plt.show()
```

```python
# ==========================================
# 3. BOX PLOT
# ==========================================
# Objective:
# To detect outliers in cholesterol values

plt.figure(figsize=(8,5))

sns.boxplot(x=df['chol'])

plt.title("Box Plot of Cholesterol")

plt.show()
```

```python
# ==========================================
# 4. SCATTER PLOT
# ==========================================
# Objective:
# To study relationship between age and cholesterol

plt.figure(figsize=(8,5))

plt.scatter(df['age'], df['chol'])

plt.title("Scatter Plot: Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")

plt.show()
```

```python
# ==========================================
# 5. SCATTER PLOT WITH BOX PLOTS
# ==========================================
# Objective:
# To visualize relationship between age and
# cholesterol along with distributions
# and outliers

sns.jointplot(
    x='age',
    y='chol',
    data=df,
    kind='scatter'
)

plt.show()

# ==========================================
# END
# ==========================================
```
