```python
# header=None because dataset has no column names
df = pd.read_csv("hepatitis.csv", header=None)

# Assign column names manually

df.columns = [
    'class', 'age', 'sex', 'steroid', 'antivirals',
    'fatigue', 'malaise', 'anorexia', 'liver_big',
    'liver_firm', 'spleen_palpable', 'spiders',
    'ascites', 'varices', 'bilirubin', 'alk_phosphate',
    'sgot', 'albumin', 'protime', 'histology'
]

# Display first rows
print("Original Dataset:\n")
print(df.head())

```

```python
# =====================================================
# a. Create data subsets for different sex
# =====================================================

# male = 1
# female = 2

male_data = df[df['sex'] == 1]

female_data = df[df['sex'] == 2]

print("\n\nMale Dataset:\n")
print(male_data.head())

print("\n\nFemale Dataset:\n")
print(female_data.head())
```

```python
# =====================================================
# b. Merge two subsets
# =====================================================

merged_data = pd.concat([male_data, female_data])

print("\n\nMerged Dataset:\n")
print(merged_data.head())
```

```python
# =====================================================
# c. Sort Data using age, sgot, protime
# =====================================================

# Convert columns to numeric if needed

df['sgot'] = pd.to_numeric(df['sgot'], errors='coerce')
df['protime'] = pd.to_numeric(df['protime'], errors='coerce')

sorted_data = df.sort_values(
    by=['age', 'sgot', 'protime'],
    ascending=[True, True, True]
)

print("\n\nSorted Dataset:\n")
print(sorted_data.head())
```

```python
# =====================================================
# d. Transposing Data
# =====================================================

transpose_data = df.transpose()

print("\n\nTransposed Data:\n")
print(transpose_data.head())
```

```python
# =====================================================
# e. Melting Data to long format
# =====================================================

melted_data = pd.melt(
    df,
    id_vars=['sex'],
    value_vars=['age', 'bilirubin', 'sgot'],
    var_name='Test_Type',
    value_name='Value'
)

print("\n\nMelted Data:\n")
print(melted_data.head())
```

```python
# =====================================================
# f. Casting data to wide format
# =====================================================

# Replace ? with NaN
melted_data['Value'] = melted_data['Value'].replace('?', pd.NA)

# Convert to numeric
melted_data['Value'] = pd.to_numeric(
    melted_data['Value'],
    errors='coerce'
)

# Create wide format table
wide_data = melted_data.pivot_table(
    index='sex',
    columns='Test_Type',
    values='Value',
    aggfunc='mean'
)

print("\nWide Format Data:\n")
print(wide_data.head())
```
