```python
# ==========================================
# HEART DISEASE DATASET VISUALIZATION
# ==========================================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ==========================================
# COLUMN NAMES
# ==========================================

columns = [
    'age', 'sex', 'cp', 'trestbps', 'chol',
    'fbs', 'restecg', 'thalach', 'exang',
    'oldpeak', 'slope', 'ca', 'thal', 'target'
]

# ==========================================
# LOAD DATASET
# ==========================================

df = pd.read_csv("Cleavland.csv", header=None, names=columns)
```

```python
# ==========================================
# DISPLAY FIRST 5 ROWS
# ==========================================

print(df.head())

# ==========================================
# DATA CLEANING
# ==========================================

# Replace ? with NaN
df.replace('?', pd.NA, inplace=True)

# Remove missing values
df.dropna(inplace=True)

# Convert all columns to numeric
for col in df.columns:
    df[col] = pd.to_numeric(df[col])
```

```python
# ==========================================
# 1. HISTOGRAM
# ==========================================
# Objective:
# To understand distribution of age

plt.figure(figsize=(8,5))

plt.hist(df['age'], bins=10)

plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
# ==========================================
# 2. DOT PLOT
# ==========================================
# Objective:
# To visualize spread of cholesterol values

plt.figure(figsize=(10,4))

sns.stripplot(x=df['chol'])

plt.title("Dot Plot of Cholesterol")

plt.show()
```

```python

# ==========================================
# 3. BAR PLOT
# ==========================================
# Objective:
# To compare chest pain type frequencies

plt.figure(figsize=(8,5))

df['cp'].value_counts().plot(kind='bar')

plt.title("Bar Plot of Chest Pain Type")
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")

plt.show()
```

```python
# ==========================================
# 4. LINE CHART
# ==========================================
# Objective:
# To study trend between age and cholesterol

df_sorted = df.sort_values(by='age')

plt.figure(figsize=(10,5))

plt.plot(df_sorted['age'], df_sorted['chol'])

plt.title("Line Chart: Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")

plt.show()
```

```python
# 1. Make a simple space for 2 side-by-side graphs
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# 2. Put the Histogram in the first graph (ax1)
sns.histplot(df['thalach'], ax=ax1, color='lightblue')

# 3. Put the Box Plot in the second graph (ax2)
sns.boxplot(x=df['thalach'], ax=ax2, color='lightgreen')

# 4. Add the Scatter Plot dots ON TOP of the Box Plot
sns.stripplot(x=df['thalach'], ax=ax2, color='red')

# 5. Show it
plt.show()
```
