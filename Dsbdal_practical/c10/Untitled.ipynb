```python
import pandas as pd

# Load dataset properly
df = pd.read_csv("Iris.csv", header=None)

# Assign correct column names (NO slicing)
df.columns = ['SepalLengthCm', 'SepalWidthCm',
              'PetalLengthCm', 'PetalWidthCm', 'Species']

df.head()
```

```python
df.columns
```

```python
df['Species'].unique()
```

```python
#create subsets for diffrenet Species
# Subset for Iris-setosa
setosa_df = df[df['Species'] == 'Iris-setosa']

# Subset for Iris-versicolor
versicolor_df = df[df['Species'] == 'Iris-versicolor']

# Display
setosa_df.head()
versicolor_df.head()
```

```python
# Merge both subsets
merged_df = pd.concat([setosa_df, versicolor_df])

merged_df.head()
```

```python
# Sort in ascending order
sorted_df = df.sort_values(by='PetalLengthCm', ascending=True)

sorted_df.head()
```

```python
# Transpose rows to columns
transposed_df = df.T

transposed_df.head()
```

```python
#melt data
# Convert wide format to long format
melted_df = pd.melt(df,
                    id_vars=['Species'],
                    value_vars=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],
                    var_name='Feature',
                    value_name='Value')

melted_df.head()
```

```python
#Pivot data
# Convert long format back to wide format
wide_df = melted_df.pivot_table(index='Species',
                               columns='Feature',
                               values='Value',
                               aggfunc='mean')

wide_df
```
