```python
import pandas as pd

# STEP 1: Load dataset
df = pd.read_csv("movies_metadata.csv", low_memory=False)

# Check columns
print(df.columns)
```

```python

print(df['original_language'].unique())
```

```python
# ----------------------------------------------------
# m. Create subsets for different languages
# ----------------------------------------------------
eng_movies = df[df['original_language'] == 'en']
fr_movies = df[df['original_language'] == 'fr']   # you can choose any language
```

```python
print("\nEnglish Movies:")
print(eng_movies.head())

print("\nFrench Movies:")
print(fr_movies.head())
```

```python
# n. Merge two subsets
# ----------------------------------------------------
merged_data = pd.concat([eng_movies, fr_movies])

print("\nMerged Data:")
print(merged_data[['title', 'original_language']].head())
```

```python
# o. Sort Data using customer ratings
# ----------------------------------------------------
sorted_data = df.sort_values(by='vote_average', ascending=False)

print("\nTop Rated Movies:")
print(sorted_data[['title', 'vote_average']].head())
```

```python
# p. Transposing Data
# ----------------------------------------------------
transposed_data = df[['title', 'vote_average', 'popularity']].head().T

print("\nTransposed Data:")
print(transposed_data)
```

```python
# q. Melting Data (Wide → Long)
# ----------------------------------------------------
melted_data = pd.melt(df,
                     id_vars=['title', 'original_language'],
                     value_vars=['vote_average', 'popularity'],
                     var_name='Metric',
                     value_name='Value')

print("\nMelted Data:")
print(melted_data.head())
```

```python
# r. Casting Data (Long → Wide)
# ----------------------------------------------------
wide_data = melted_data.pivot_table(index=['title', 'original_language'],
                                    columns='Metric',
                                    values='Value',
                                    aggfunc='mean')

print("\nWide Data:")
print(wide_data.head())
```
